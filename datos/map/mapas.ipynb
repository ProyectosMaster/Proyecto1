```python
# df_lugares = pd.read_excel("../../datos/raw_data/df_location.xlsx")
df_lugares = pd.read_excel("df_location.xlsx")
```

```python
Espania = [40.463669,-3.749220]

latitudes = df_lugares['Latitud']
longitudes = df_lugares['Longitud']
centroide = [latitudes.mean(), longitudes.mean()]

Map = folium.Map(Espania,zoom_start=11, min_zoom=11, max_zoom=11
                 ,tiles="Cartodb Positron", )

Map
```

```python
for i in range(0, len(df_lugares)):
    folium.Marker([df_lugares.iloc[i]['Latitud'], df_lugares.iloc[i]['Longitud']],
                  popup = df_lugares.iloc[i]['Cliente']).add_to(Map)
    
Map
```

```python
from folium import Icon
```

```python
for i in range(0, len(df_lugares)):
    lat = df_lugares.iloc[i]['Latitud']
    lon = df_lugares.iloc[i]['Longitud']
    cliente = df_lugares.iloc[i]['Cliente']
    
    # Verificar si el cliente es 'Almacen'
    if cliente == 'Almacén':
        # Cambiar el ícono y el color del marcador
        icono = Icon(color='red', icon='glyphicon glyphicon-home', prefix='glyphicon')
    else:
        # Marcador por defecto para otros clientes
        icono = Icon(color='blue', icon='glyphicon glyphicon-map-marker', prefix='glyphicon')
    
    # Crear el marcador con el ícono personalizado
    folium.Marker(
        location=[lat, lon],
        popup=cliente,
        icon=icono
    ).add_to(Map)
    
Map
```
