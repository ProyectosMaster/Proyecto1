```python
import numpy as np
from pulp import LpProblem, LpVariable, lpSum, LpMinimize, value

# Datos iniciales
matriz_distancias = np.array([
    [0, 7.5625, 15.5365, 1.1998, 4.7145, 3.6614],  # Cliente 0 a todos
    [7.5625, 0, 3.3838, 7.7433, 14.572, 10.7361],  # Cliente 1 a todos
    [15.5365, 3.3838, 0, 12.5438, 0, 13.9021],     # Cliente 2 a todos
    [1.1998, 7.7433, 12.5438, 0, 5.0721, 3.3673],  # Cliente 3 a todos
    [4.7145, 14.572, 0, 5.0721, 0, 4.5417],        # Cliente 4 a todos
    [3.6614, 10.7361, 13.9021, 3.3673, 4.5417, 0], # Almacén a todos
])

vehiculos = [
    {"id": 1, "capacidad": 2026, "costo_km": 0.2, "autonomia": 603},
    {"id": 2, "capacidad": 4362, "costo_km": 0.14, "autonomia": 630},
    {"id": 3, "capacidad": 4881, "costo_km": 0.2, "autonomia": 664},
    {"id": 4, "capacidad": 3321, "costo_km": 0.19, "autonomia": 514},
    {"id": 5, "capacidad": 10000, "costo_km": 0.32, "autonomia": 350},
    {"id": 6, "capacidad": 3129, "costo_km": 0.14, "autonomia": 791},
]

demandas = [909, 959, 960, 980, 979]

# Cantidad de nodos
n_clientes = len(demandas)
n_nodos = n_clientes + 1  # Clientes + almacén

# Crear el modelo
modelo = LpProblem("Optimizar_Rutas", LpMinimize)

# Variables de decisión
x = LpVariable.dicts("x", [(i, j, v["id"]) for i in range(n_nodos) for j in range(n_nodos) for v in vehiculos if i != j],
                     cat="Binary")
y = LpVariable.dicts("y", [(i, v["id"]) for i in range(n_nodos) for v in vehiculos], lowBound=0, cat="Continuous")

# Función objetivo: minimizar el costo total
costo_total = lpSum(x[(i, j, v["id"])] * matriz_distancias[i][j] * v["costo_km"]
                    for i in range(n_nodos) for j in range(n_nodos) for v in vehiculos if i != j)
modelo += costo_total

# Restricciones
# 1. Cada cliente debe ser atendido exactamente una vez
for i in range(1, n_clientes + 1):
    modelo += lpSum(x[(i, j, v["id"])] for j in range(n_nodos) for v in vehiculos if i != j) == 1
    
# 6. Rutas deben comenzar y terminar en el almacén
for v in vehiculos:
    modelo += lpSum(x[(n_clientes, j, v["id"])] for j in range(n_nodos) if n_clientes != j) == 1
    modelo += lpSum(x[(i, n_clientes, v["id"])] for i in range(n_nodos) if i != n_clientes) == 1

# 2. Salida e ingreso balanceados por cliente
for v in vehiculos:
    for i in range(n_nodos):
        modelo += lpSum(x[(i, j, v["id"])] for j in range(n_nodos) if i != j) == lpSum(x[(j, i, v["id"])] for j in range(n_nodos) if i != j)

# 3. Capacidad del vehículo no debe ser excedida
for v in vehiculos:
    modelo += lpSum(y[(i, v["id"])] for i in range(1, n_clientes + 1)) <= v["capacidad"]

# 4. Demanda de cada cliente debe ser cumplida
for v in vehiculos:
    for i in range(1, n_clientes + 1):
        modelo += y[(i, v["id"])] <= demandas[i - 1]
        modelo += lpSum(x[(i, j, v["id"])] for j in range(n_nodos) if i != j) == lpSum(x[(j, i, v["id"])] for j in range(n_nodos) if i != j)

# 5. Respetar la autonomía de los vehículos
for v in vehiculos:
    modelo += lpSum(x[(i, j, v["id"])] * matriz_distancias[i][j] for i in range(n_nodos) for j in range(n_nodos) if i != j) <= v["autonomia"]



# Resolver el modelo
modelo.solve()

# Mostrar resultados
print("Estado del modelo:",  lp.LpStatus[modelo.status])
print("Costo total mínimo:", value(modelo.objective))

# Rutas asignadas
for v in vehiculos:
    print(f"Vehículo {v['id']}:")
    for i in range(n_nodos):
        for j in range(n_nodos):
            if i != j and x[(i, j, v['id'])].varValue > 0.5:
                print(f"De {i} a {j}")
```
